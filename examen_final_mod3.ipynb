```python
# Creo una funcion para Exploración de los valores unicos del DataFrame

def explorar_valores_unicos(df):
    for columna in df.columns:
        valores_unicos=df[columna].unique()
        total_unicos=len(valores_unicos)

        print(f"Columna : {columna}")
        print(f"Los valores unicos son: {valores_unicos}")
        print(f"Total valores unicos : {total_unicos}")
        print("---------------------------")

#Llamo a la funcion para Exploracion de valores unicos de flight
explorar_valores_unicos(df_flight)
```

```python
# Llamo a la funcion para Exploración de los valores unicos de loyalty

explorar_valores_unicos(df_loyalty)
```

```python
#segun la compañia Salary es una estimacion del salario de los subcriptores
#aqui calculo la estadistica de la columna, uso (.T) transpone el resultado en filas,(reset.index)restablezco el indice para convertir en columnas,y (round)redondeo los valores numericos a los enteros.
df_loyalty['Salary'].describe().T.reset_index().round()
```

```python
#observo los valos negativos de salary
df_loyalty.loc[df_loyalty['Salary'] < 0, 'Salary']
```

```python
#convierto los valores negativos a positivos de la columna salary, usando el metodo np.where() que me permite convertir un array de salarios en positivos.
df_loyalty['Salary'] = np.where(df_loyalty['Salary'] < 0, -df_loyalty['Salary'], df_loyalty['Salary'])
```

```python
#compruebo que se haya modificado el valor a positivo de la columna salary y observo que efectivamente ya no hay valores negativos.
df_loyalty.loc[df_loyalty['Salary'] < 0, 'Salary']
```

```python
# Decido unir los datos mediante Loyalty number que es la columna en comun 
# uso el parametro how para determinar el tipo de union en este caso escogemos tipo de unión de INNER JOIN-

df_unido = df_loyalty.merge(df_flight, on=["Loyalty Number"], how="inner")
```
